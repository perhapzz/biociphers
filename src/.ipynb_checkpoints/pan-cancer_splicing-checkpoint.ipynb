```python
def write_pval(pval, x, y, threshold):
    if pval < threshold:
        plt.text(x, y, '%.2e' % Decimal(pval), fontsize=font_size, ha='center', va='bottom')
    else:
        plt.text(x, y, '%.3f' % pval, fontsize=font_size, ha='center', va='bottom')
```

## selection of high-attribution features

```python
## plot distribution of attribution values showing median to find kneepoints


## function to smoothen curve made up of attribution values
def smooth(y, box_pts):
        y_smooth = np.convolve(y, np.ones((box_pts,)) / float(box_pts), mode='valid')
        return y_smooth
    

attr_median = pd.read_csv(dataPval_file,
                        sep='\t', index_col=0)


attr_median.index = attr_median.index.str.split(':').str[1].map(dct_ref) + ';' + attr_median.index

attr_median['pval'] = -np.log10(attr_median['Corrected P-values (FDR <= 1%)'])

## use absolute attribution values because direction of attrubiton has no meaning for splicing data
attr_median['Median-Attributions'] = abs(attr_median['Median-Attributions'])


## sort df_attr from highest to lowest attribution value
attr_median = attr_median.sort_values(by='Median-Attributions', ascending=False)

attr_median_pos = attr_median[attr_median['Median-Attributions'] >= 0]


x = [x for x in range(len(attr_median.index))]
xpos = [x for x in range(len(attr_median_pos))]


y = smooth(attr_median_pos['Median-Attributions'], 25)
smooth_x = np.arange(len(y))

kneedle = KneeLocator(smooth_x, y, S=1.0, curve="convex", direction="decreasing")

cutoff_pos = y[kneedle.knee]



## plot distribution curve for attribution values and show knee points
red1 = [x for x in attr_median['Median-Attributions'] if x > cutoff_pos]
black = [x for x in attr_median['Median-Attributions'] if x <= cutoff_pos]

fig,ax = plt.subplots(figsize=(20,10))

font_size = 28

plt.title('Ranking of splice junctions by median attribution value (Cancer-wide)', fontsize=font_size)

plt.scatter([x for x in range(len(red1))], red1, s=1, color='red', rasterized=True)
plt.scatter([x+len(red1) for x in range(len(black))], black, s=1, color='grey', rasterized=True)


plt.xticks(fontsize = font_size)
plt.yticks(fontsize = font_size)
plt.xlabel('Splice junction rank', fontsize=font_size)
plt.ylabel('Median attribution value', fontsize=font_size)
plt.xlim(min(x), max(x))
plt.axhline(cutoff_pos, linestyle='--', color='red')

plt.text(max(x), max(attr_median['Median-Attributions']), str(len(red1)) + ' splice junctions above positive knee-point \n',# +\
        fontsize=font_size,
        ha='right', va='top')

plt.savefig('../figures/attribution_distribution_splicing_median.pdf', dpi=600, bbox_inches='tight')
plt.show()
plt.close()
```

```python
## make volcano plots using p-values associated with attribution values
pval_cutoff = 3

df_volcano_significant = attr_median[attr_median['pval'] > pval_cutoff]
df_volcano_pos = df_volcano_significant[df_volcano_significant['Median-Attributions'] > cutoff_pos]

df_volcano_nonsignificant = attr_median[~attr_median.index.isin(df_volcano_pos.index.tolist())]


## export  with significant high positive or negative attribution values for GO and other downstream analyses
_high_pos = df_volcano_pos.index

## making neutral set by selecting similar number of neutral  as the size of the high lsv population at the center of the distribution
num_neutral = len(_high_pos) ## number of features included in neutral set is equal to set that has most features between high-negative and high-positive
attr_median_neutral = attr_median[attr_median['pval'] < 1.3]
attr_median_neutral = attr_median_neutral[~attr_median_neutral.index.str.startswith('AC0')] ## remove weird 
_neutral = attr_median_neutral.index.values[-num_neutral:]
df_volcano_neutral = attr_median[attr_median.index.isin(_neutral)]


fig,ax = plt.subplots(figsize=(10,10))

font_size=32

plt.title('Splice junction usage', fontsize=font_size)
plt.plot(df_volcano_pos['Median-Attributions'], df_volcano_pos['pval'], 'o', color='lightcoral', alpha=0.5, rasterized=True)
plt.plot(df_volcano_nonsignificant['Median-Attributions'], df_volcano_nonsignificant['pval'], 'o', color='grey', alpha=0.5, rasterized=True)
plt.plot(df_volcano_neutral['Median-Attributions'], df_volcano_neutral['pval'], 'o', color='purple', alpha=0.5, rasterized=True)

plt.text(0.0075, 275, 'n=' + str(len(_high_pos)), ha='right', va='top', fontsize=font_size)
plt.text(0, 310, 'n=' + str(len(_neutral)), ha='left', va='bottom', fontsize=font_size)
plt.plot((0.0001,0.0005), (295, 310), color='black')

plt.axvline(cutoff_pos, linestyle='--', color='black')
plt.axhline(4, linestyle='--', color='black')

plt.xticks(fontsize=font_size, rotation=45, ha='right', va='top')
plt.yticks(fontsize=font_size)

plt.xlabel('Attribution values', fontsize=font_size)
plt.ylabel('-log10(p-values)', fontsize=font_size)

plt.xlim(0, 0.008)
plt.ylim(0, 325)

plt.savefig('../figures/volcano_attribution_splicing_median.pdf',
            dpi=600, bbox_inches='tight')
plt.show()
plt.close()
```

```python
## Plot how attribution values of a set of genes look across cancer types

def plot_top_genes_perTissue(gene_list, groups, dct_tumor, df_pc, title, colors, save_as):
    print ('top genes is the number of top genes per tissue to plot')
    
    
    tumors = [x for x in set(dct_tumor.values()) if x not in ['stomach_cancer_tumor', 'ovarian_cancer_tumor']] ## remove stomach and ovarian cancer because we have low numbers
    
    
    tumor_samples = [x for x in df_pc.index if dct_tumor[x] in tumors]
    
    
    df = df_pc[df_pc.index.isin(tumor_samples)]
    
    
    fig,ax = plt.subplots(figsize=(12.5,10))
    
    font_size = 38
    
    starting_x = 0
    
    
    for ii,genes in enumerate(gene_list):
        
        genes = [gene for gene in genes if gene in df.columns]
        
        df_tmp = df[genes]
        
        for tumor_type in tumors:
            samples = [x for x in tumor_samples if dct_tumor[x] == tumor_type]
            median = df_tmp[df_tmp.index.isin(samples)].median(axis=0)
            
            df_tmp.loc['median_'+tumor_type] = median
            
        df_tmp = df_tmp[df_tmp.index.str.contains('median')]
        
        for iii,gene in enumerate(genes):
            x = [iii for x in range(len(df_tmp))]
            y = df_tmp[gene].tolist()
            
            plt.plot(x, y, 'o', color = colors[ii], rasterized=True)
        
        starting_x += len(genes)
        
    plt.ylabel('Median attribution value\nper tumor type', fontsize=font_size)
    plt.xlabel('Splice junctions', fontsize=font_size)
    plt.yticks(fontsize=font_size)
    plt.xticks(fontsize=font_size)
    plt.xlim(-1, 100)
    plt.ylim(0, 0.009)
    
    ## manual legend
    patches = []
    for ii,group in enumerate(groups):
        patch = mpatches.Patch(color=colors[ii], label=group, alpha=0.6)
        patches.append(patch)
        
    plt.legend(handles=patches, frameon=False, fontsize=font_size, labelspacing = 0.2)
    
    plt.savefig(save_as + '.pdf', dpi=600, bbox_inches='tight')
    
    plt.show()
    plt.close()
    

## plotting median attribution values of splice junctions in TSGs and oncogenes from COSMIC
## only looking at junction with highest attribution (if any) found for each COSMIC gene
num_genes = 100

df_cosmic = pd.read_csv(cosmic_file,
                       sep='\t', index_col='Gene Symbol')

df_cosmic = df_cosmic[df_cosmic['Role in Cancer'].notna()]

TSG = df_cosmic[df_cosmic['Role in Cancer'].str.contains('TSG')]
TSG = TSG[~TSG['Role in Cancer'].str.contains('oncogene')]
TSG = TSG.index.tolist()

oncogene = df_cosmic[df_cosmic['Role in Cancer'].str.contains('oncogene')]
oncogene = oncogene[~oncogene['Role in Cancer'].str.contains('TSG')]
oncogene = oncogene.index.tolist()


TSG_junctions = []
oncogene_junctions = []
for junction in attr_median.index:
    gene = junction.split(';')[0]
    if gene in TSG:
        TSG_junctions.append(junction)
    elif gene in oncogene:
        oncogene_junctions.append(junction)

TSG_junctions = TSG_junctions[:num_genes]
oncogene_junctions = oncogene_junctions[:num_genes]


title = 'Splice junctions in COSMIC genes'
groups = ['Tumor-suppressor genes', 'Oncogenes']
colors = ['limegreen', 'orange']
save_as = '../figures/attribution_median_tumours_COSMIC_junctions'

plot_top_genes_perTissue([TSG_junctions, oncogene_junctions], groups, dct_tumor, df_pc, title, colors, save_as)



## plotting median attribution values of high-attribution splice junctions
genes_enriched = pd.read_csv(lsvHigh_file,
                            header=None)[0].tolist()
genes_enriched = [x for x in genes_enriched if x not in oncogene and x not in TSG]
genes_enriched = genes_enriched[:num_genes]


genes_neutral = pd.read_csv(lsvNeutral_file,
                            header=None)[0].tolist()
genes_neutral = [x for x in genes_neutral if x not in oncogene and x not in TSG]
genes_neutral = genes_neutral[int(len(genes_neutral)/2)-int(num_genes/2): int(len(genes_neutral)/2)-int(num_genes/2)+num_genes]


colors = ['grey', 'lightcoral']

title = 'Top ' + str(num_genes) + ' splice junctions'
groups = ['Splice junctions neutral', 'Splice junctions positive']
save_as = '../figures/attribution_lsv_median_tumours'

plot_top_genes_perTissue([genes_neutral, genes_enriched], groups, dct_tumor, df_pc, title, colors, save_as)



```

## assess evolutionary conservation

```python
## computing convervation scores

def assess_conservation(conservation_file, dct_enstToensg):
    
    header = ['query acc', 'query length', 'subject acc', 'subject length', 'bit score', 'score',
          'alignment length', '% identity', 'subject tax id']
    
    df_conservation = pd.read_csv(conservation_file, header=None, sep='\t', comment='#')
    
    df_conservation.columns = header
    
    df_conservation['query acc'] = df_conservation['query acc'].map(dct_enstToensg)
    
    
    genes = set(df_conservation['query acc'].tolist())

    dct_conservation = {}
    for ii,gene in enumerate(genes):
        ## subset dataframe to one gene
        df_tmp = df_conservation[df_conservation['query acc'] == gene]
        ## remove human taxid because normalized score (see 4 lines below) will always be 1.847
        df_tmp = df_tmp[df_tmp['subject tax id'] != 9606]
        ## order dataframe by descending bit score values
        df_tmp = df_tmp.sort_values(by='bit score', axis=0, ascending=False)
        ## select match with greatest bit score for each taxid
        df_tmp.index = df_tmp['subject tax id']
        df_tmp = df_tmp[~df_tmp.index.duplicated(keep='first')]
        ## add normalized score corresponding to bit score normalized by query length
        df_tmp['normalized score'] = df_tmp['bit score'] / df_tmp['query length']

        normalized_score = np.sum(df_tmp['normalized score'])
        try:
            dct_conservation[df_tmp.iloc[0]['query acc']] = normalized_score
        except IndexError:
            ## this exception catches genes with no match in species other than Homo sapiens and assigns a score of 0
            dct_conservation[gene] = 0
            
    return dct_conservation
        
    
with open(dctEnstEnsg_file, 'rb') as f:
    dct_enstToensg = pickle.load(f)
    
    
## high LSVs
dct_conservation_high = assess_conservation(conservationPos_file, dct_enstToensg)


## neutral genes
dct_conservation_neutral = assess_conservation(conservationNeutral_file, dct_enstToensg)


print ('ready to plot conservation scores')
```

```python
## plot conservation scores

groups = ['Neutral\n(n=' + str(len(dct_conservation_neutral)) + ')',
          'Positive\n(n=' + str(len(dct_conservation_high)) + ')']

dcts = [dct_conservation_neutral, dct_conservation_high]

fig,ax = plt.subplots(figsize=(7.5,10))

font_size = 32

plt.title('Evolutionary conservation in genes\nwith variable splice junctions', fontsize=font_size)

tstat_neutralhigh, pval_neutralhigh = stats.ttest_ind([x for x in dct_conservation_neutral.values()],
                                                              [x for x in dct_conservation_high.values()])


for ii,group in enumerate(groups):
    
    y = dcts[ii].values()
    x = np.random.normal(ii, 0.15, size=len(y))
#     plt.scatter(x, y, color = ['grey', 'lightcoral', 'dodgerblue'][ii], alpha = 0.5)

    vp = plt.violinplot(
            y, positions = [ii],
        showmeans=False, showmedians=False,
            showextrema=False)

    for pc in vp['bodies']:
        pc.set_facecolor(['lightgrey', 'lightcoral', 'dodgerblue'][ii])
        pc.set_edgecolor('black')
        pc.set_alpha(1)
        
    box1 = plt.boxplot(y, positions = [ii], showfliers=False)
    plt.setp(box1['medians'], color='black', linewidth=2)
    
    
plt.plot([0,1], [8.75,8.75], color='black')

write_pval(pval_neutralhigh, 0.5, 8.75, 0.001)
    
plt.xticks(range(len(groups)), groups, fontsize=font_size)
plt.xlabel('Junction category', fontsize=font_size)
plt.yticks(fontsize=font_size)
plt.ylabel('Conservation score\n(length-normalized BLAST bit score)', fontsize=font_size)
plt.xlim(-0.6, 1.6)
plt.ylim(0,10.5)

plt.show()
plt.close()
```

```python
## Plot LOEUF scores (gnomAD) for enriched and neutral genes

## fetch LOEUF score

df_loeuf = pd.read_csv(loeuf_file,
                      sep='\t', index_col='gene')

## only keep transcript with highest LOEUF score gene a gene has more than one transcript listed in gnomAD
df_loeuf = df_loeuf[~df_loeuf.index.duplicated(keep='first')]

## remove genes for which no LOEUF score is defined (nan)
df_loeuf = df_loeuf[df_loeuf['oe_lof_upper'].notna()]


genes_enriched = pd.read_csv(lsvHigh_file,
                           header=None)[0].tolist()
genes_neutral = pd.read_csv(lsvNeutral_file,
                           header=None)[0].tolist()


groups = ['Neutral\n(n=' + str(len(genes_neutral)) + ')',
          'Positive\n(n=' + str(len(genes_enriched)) + ')']

dct_loeuf_neutral = {}
dct_loeuf_enriched = {}

for gene in set([x.split(';')[0] for x in genes_neutral]):
    try:
        dct_loeuf_neutral[gene] = df_loeuf['oe_lof_upper'][gene]
    except KeyError:
        continue
    
for gene in set([x.split(';')[0] for x in genes_enriched]):
    try:
        dct_loeuf_enriched[gene] = df_loeuf['oe_lof_upper'][gene]
    except KeyError:
        continue
    
    
dcts = [dct_loeuf_neutral, dct_loeuf_enriched]


fig,ax = plt.subplots(figsize=(7.5,10))

font_size = 32

plt.title('Loss of function mutation frequency\nin genes with variable splice junctions', fontsize=font_size)

tstat_neutralEnriched, pval_neutralEnriched = stats.ttest_ind([x for x in dct_loeuf_neutral.values()],
                                                              [x for x in dct_loeuf_enriched.values()])


for ii,group in enumerate(groups):
    
    y = dcts[ii].values()
    x = np.random.normal(ii, 0.15, size=len(y))
#     plt.scatter(x, y, color = ['grey', 'lightcoral', 'dodgerblue'][ii], alpha = 0.5)

    vp = plt.violinplot(
            y, positions = [ii],
        showmeans=False, showmedians=False,
            showextrema=False)
    
    for pc in vp['bodies']:
        pc.set_facecolor(['lightgrey', 'lightcoral', 'dodgerblue'][ii])
        pc.set_edgecolor('black')
        pc.set_alpha(1)
        
    box1 = plt.boxplot(y, positions = [ii], showfliers=False)
    plt.setp(box1['medians'], color='black', linewidth=2)
    
    
plt.plot([0,1], [2.1,2.1], color='black')

write_pval(pval_neutralEnriched, 0.5, 2.1, 0.001)

plt.xticks(range(len(groups)), groups, fontsize=font_size)
plt.xlabel('Junction category', fontsize=font_size)
plt.yticks(fontsize=font_size)
plt.ylabel('gnomAD LOEUF score', fontsize=font_size)
plt.xlim(-0.6, 1.6)
plt.ylim(0,2.3)

plt.savefig('../figures/loeuf_score_lsv_genes.pdf',
            dpi=600, bbox_inches='tight')
plt.show()
plt.close()

```

```python
## Plot gene lengths
def make_transcriptLength_dct(infile):
    
    dct_length = {}
    with open(infile, 'r') as infile:
        
        for ii,line in enumerate(infile):
            line = line.strip()
            
            if '>' in line:
                gene_id = line.split('gene_symbol:')[1].split(' ')[0]
                count = 0
                
            else:
                count += len(line)
                dct_length[gene_id] = count
                
    return dct_length
                
infile = lsvHigh_fasta
dct_length_high = make_transcriptLength_dct(infile)

infile = lsvNeutral_fasta
dct_length_neutral = make_transcriptLength_dct(infile)
    
    
dcts_lengths = [dct_length_neutral, dct_length_high]


groups = ['Neutral\n(n=' + str(len(dct_length_neutral)) + ')',
          'Positive\n(n=' + str(len(dct_length_high)) + ')']

## plot lncRNA gene lengths
fig,ax = plt.subplots(figsize=(7.5,10))

font_size = 32

plt.title('Length of \n with variable splice junctions', fontsize = font_size)

tstat_neutralhigh, pval_neutralhigh = stats.ttest_ind([dct_length_neutral[x] for x in dct_length_neutral],
                                                              [dct_length_high[x] for x in dct_length_high])

for ii,group in enumerate(groups):
    
    y = [dcts_lengths[ii][x] for x in dcts_lengths[ii]]
    x = np.random.normal(ii, 0.15, size=len(y))
    
    vp = plt.violinplot(
            y, positions = [ii],
        showmeans=False, showmedians=False,
            showextrema=False)

    for pc in vp['bodies']:
        pc.set_facecolor(['lightgrey', 'lightcoral', 'dodgerblue'][ii])
        pc.set_edgecolor('black')
        pc.set_alpha(1)
        
    box1 = plt.boxplot(y, positions = [ii], showfliers=False)
    plt.setp(box1['medians'], color='black', linewidth=2)
    
    
plt.plot([0,1], [17500, 17500], color='black')

write_pval(pval_neutralhigh, 0.5, 17500, 0.001)

plt.xticks(range(len(groups)), groups, fontsize=font_size)
plt.xlabel('Gene category', fontsize=font_size)
plt.yticks(fontsize=font_size)
plt.ylabel('Coding sequence length (bp)', fontsize=font_size)
plt.xlim(-0.6, 1.6)
plt.ylim(0, 21000)

plt.savefig('../figures/gene_lsv_transcriptLength.pdf',
            dpi=600, bbox_inches='tight')
plt.show()
plt.close()
```

```python
## plot disorderness level around splice junctions high in cancer

dct_high = {}
with open(disordernessHigh_file, 'r') as infile:
    
    for ii,line in enumerate(infile):
        line = line.strip()
        if '>' in line:
            lsvID = line.strip('>')
            dct_high[lsvID] = {}
            
        else:
            if '#' not in line:
                line = line.split('\t')
                if len(line) > 1:
                    dct_high[lsvID][line[0]] = float(line[2])
                    
                    
dct_neutral = {}
with open(disordernessNeutral_file, 'r') as infile:
    
    for ii,line in enumerate(infile):
        line = line.strip()
        if '>' in line:
            lsvID = line.strip('>')
            dct_neutral[lsvID] = {}
            
        else:
            if '#' not in line:
                line = line.split('\t')
                if len(line) > 1:
                    dct_neutral[lsvID][line[0]] = float(line[2])
        
            
median_high = [np.median(list(dct_high[x].values())) for x in dct_high]
median_neutral = [np.median(list(dct_neutral[x].values())) for x in dct_neutral]


## function to change boxplot colors
def box_plot(data, edge_color, fill_color):
    bp = ax.boxplot(data, patch_artist=True)
    
    for element in ['boxes', 'whiskers', 'fliers', 'means', 'medians', 'caps']:
        plt.setp(bp[element], color=edge_color)

    for patch in bp['boxes']:
        patch.set(facecolor=fill_color)
        
fig,ax = plt.subplots(figsize=(5,10))


tstat_disorder, pval_disorder = stats.ttest_ind(median_neutral, median_high)

font_size = 24

plt.title('Level of disorderness around\nsplice junctions\nhigh in cancer', fontsize=font_size)

violin = ax.violinplot([median_neutral, median_high], showextrema=False)
box_plot([median_neutral, median_high], 'black', 'none')
# bp = plt.boxplot([median_neutral, median_high], patch_artist=True)


for ii,pc in enumerate(violin['bodies']):
    pc.set_facecolor(['lightgrey', 'lightcoral'][ii])
    pc.set_edgecolor('black')
    pc.set_alpha(1)
    
    
plt.xticks([1,2], ['Neutral\nn=' + str(len(dct_neutral)), 'high\nn=' + str(len(dct_high))], fontsize=font_size)
plt.xlabel('Junction category', fontsize=font_size)
plt.yticks(fontsize=font_size)
plt.ylabel('Protein disorderness\n(IUPred score)', fontsize=font_size)
plt.ylim(0, 1.1)


plt.plot([1,2], [1.02, 1.02], color='black')

if pval_disorder < 0.001:
    plt.text(1.5, 1.02,
             '%.2e' % Decimal(pval_disorder), fontsize=font_size, ha='center', va='bottom')
else:
    plt.text(1.5, 1.02,
             '%.3f' % pval_disorder, fontsize=font_size, ha='center', va='bottom')

plt.savefig('../figures/LSV_proteinDisorderness.pdf',
            dpi=600, bbox_inches='tight')
plt.show()
plt.close()
```
